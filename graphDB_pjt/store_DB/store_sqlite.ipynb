```python
def save_df_to_sqlite(db_path: str, df: pd.DataFrame, table_name: str):
    with sqlite3.connect(db_path) as conn:
        df.to_sql(table_name, conn, if_exists="replace", index=False)
        print(f"{table_name} 테이블 저장 완료 : {len(df)} rows")
```

```python
save_df_to_sqlite(db_path, stpo_df, "stpo")
save_df_to_sqlite(db_path, mast_df, "mast")
save_df_to_sqlite(db_path, eina_df, "eina")
```
